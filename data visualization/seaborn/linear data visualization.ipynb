```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline

sns.set_style("darkgrid")
sns.set_context("paper")
# 设置风格、尺度

import warnings
warnings.filterwarnings('ignore') 
# 不发出警告
```

```python
# 基本用法

tips = sns.load_dataset('tips')       # 加载数据
print(tips.head())

sns.lmplot(x='total_bill',y='tip',
           hue='smoker',
           data=tips,
           palette='Set1',
           ci=70,                     # 误差值
           size=5,                    # 图表大小
           markers=['+','o'])         # 点样式
```

```python
# 拆分多个表格

sns.lmplot(x='total_bill',y='tip',
           col='smoker',
           data=tips)
```

```python
# 多图表1

sns.lmplot(x='size',y='total_bill',
           hue='day',col='day',
           data=tips,
           aspect=.6,      # 长宽比
           x_jitter=.3,    # 给x或者y轴随机增加噪音点
           col_wrap=4)     # 每行的列数
```

```python
# 多图表2

sns.lmplot(x='total_bill',y='tip',   # x轴total_bill, y轴tip
           row='sex',col='time',     # 行为sex字段，列为time字段
           data=tips,
           size=4)
```

```python
# 非线性回归

sns.lmplot(x='total_bill',y='tip',data=tips,
           order=2)                         # 阶数
```
